```python
#lib and dependencies
%matplotlib inline
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#loads the data from our csv file to our main dataframe
ogdata = pd.read_csv('data/athlete_events.csv')
```

```python
ogdata.info()
ogdata.dtypes
```

```python
ogdata.iloc[33]
ogdata.columns
ogdata.head()
```

```python
gold_med = ogdata[ogdata['Medal'] == 'Gold']
silver_med = ogdata[ogdata['Medal'] == 'Silver']
bronze_med = ogdata[ogdata['Medal'] == 'Bronze']
```

```python
gold_med.info()
silver_med.info()
bronze_med.info()
```

```python
medalists = gold_med.append(silver_med)
medalists.tail()
```

```python
medalists = medalists.append(bronze_med)
medalists.tail()
```

```python
medalists.describe()
```
